```python
import pandas as pd
```

```python
df = pd.read_csv("obama2013.csv")
df['PublishedAt'] = pd.to_datetime(df['PublishedAt'])
df.describe()
```

```python
#inspect distribution of the year
df["PublishedAt"].dt.year.value_counts()
```

```python
#only include comments from certain year
#remove /n from comments
df = df[(df['PublishedAt'].dt.year >= 2013) & (df['PublishedAt'].dt.year <= 2017)]
df.reset_index(drop=True, inplace=True)
df.loc[:,'Comment'] = df['Comment'].str.replace('\n', ' ')
df
```

```python
df.to_csv('obama2013_preprocessed.csv', index=False)
```
