```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import accelerate
device = "cuda" # the device to load the model onto

model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen1.5-14B-Chat-GPTQ-Int4",device_map="auto",trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen1.5-14B-Chat-GPTQ-Int4",trust_remote_code=True)


```

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/year 3/thesis/obama2013_preprocessed.csv')
df
```

```python
df.Comment[5]
```

```python
all_prompt = "Answer with one number. What is the political polarity score for the comment on a scale from -1 to 1? : I love this candidate so much, they are the best. They will solve all of our problems in our country => 0.9. Answer with one number. What is the political polarity score for the comment on a scale from -1 to 1? : I am supportive of both sides. All I want is a respectul election process => 0. Answer with one number. What is the political polarity score for the comment on a scale from -1 to 1? : I hate this candiate so much, they are the worst. They will worsen all of our problems in our country. => -0.9 Answer with one number. What is the political polarity score for the comment on a scale from -1 to 1? : "
prompt = " 👉✌️❤️🌺🙏✝️🙏💖💔🌞🌟🌎🍀💝💔👍👉=>"
messages = [
    {"role": "system", "content": "you are an expert political scientist who only answers in numbers. If you cannot answer, return 100."},
    {"role": "user", "content": all_prompt + prompt}
]
text = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True
)
model_inputs = tokenizer([text], return_tensors="pt").to(device)

generated_ids = model.generate(
    model_inputs.input_ids,
    max_new_tokens=32
)
generated_ids = [
    output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
]

response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
print(response)
```

```python
from tqdm import tqdm
polarity_index = []
for row in tqdm(df.Comment):
  prompt = row + "=>"
  all_prompt = "Answer with one number. What is the political polarity score for the comment on a scale from -1 to 1? : I love this candidate so much, they are the best. They will solve all of our problems in our country => 0.9. Answer with one number. What is the political polarity score for the comment on a scale from -1 to 1? : I am supportive of both sides. All I want is a respectful election process => 0. Answer with one number. What is the political polarity score for the comment on a scale from -1 to 1? : I hate this candidate so much, they are the worst. They will worsen all of the problems in our country. => -0.9. Answer with one number. What is the political polarity score for the comment on a scale from -1 to 1? : "
  messages = [
    {"role": "system", "content": "You are a leading political science expert who only answers in numbers. If you cannot answer, return 100."},
    {"role": "user", "content": all_prompt + prompt}
  ]
  text = tokenizer.apply_chat_template(
      messages,
      tokenize=False,
      add_generation_prompt=True
  )
  model_inputs = tokenizer([text], return_tensors="pt").to(device)

  generated_ids = model.generate(
      model_inputs.input_ids,
      max_new_tokens=32
  )
  generated_ids = [
      output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
  ]

  response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
  polarity_index.append(response)

df["Polarity_index2"] = polarity_index
```

```python
for i,j in zip(df.Polarity_index, df.Comment):
  print(i,j)
```
